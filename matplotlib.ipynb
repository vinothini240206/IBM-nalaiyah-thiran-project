```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("x values") #to define x axis names
plt.ylabel("y values") # ro define y axis name
plt.title("this is my 1st graph") # to give the title of the graph
```

```python
plt.plot(x,y,'r')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("this is my 1st graph")
```

```python
plt.plot(x,y,'r--')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("this is my 1st graph")
```

```python
plt.plot(x,y,'g*')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("this is my 1st graph")
```

```python
plt.plot(x,y,'g-.*')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("this is my 1st graph")
```

**subplots**

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("this is my 1st graph")

plt.subplot(2,2,4)
plt.plot(x,y,'g*')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("this is my 1st graph")
```
